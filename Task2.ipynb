```python
import pandas as pd

df = pd.read_csv("Titanic-Dataset.csv")
print('Data loaded successfully.')

# Display the first few rows of the dataframe and the data types of each column
display(df.head())
print(df.dtypes)
```

```python
# Import necessary libraries
import pandas as pd


# Display the structure of the dataset
print(df.head())

# Check the shape of the dataset
print(df.shape)
```

```python
# Check for missing values in each column
df_missing = df.isnull().sum()
print('Missing values in each column:')
print(df_missing)

# Suggest handling methods for missing values
missing_handling_suggestions = 'For Age, we can impute missing values using the median age. For Cabin and Embarked, we might consider filling missing values with a placeholder like "Unknown" or removing the rows if the percentage of missing data is too high.'
print(missing_handling_suggestions)
```

```python
# Impute missing values for 'Age' with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Check if the imputation was successful
df_missing_post_imputation = df.isnull().sum()
print('Missing values after imputation for Age:')
print(df_missing_post_imputation)
```

```python
# Fill missing values for 'Cabin' and 'Embarked' with the most common value (mode)
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Check if the imputation was successful
df_missing_post_all_imputation = df.isnull().sum()
print('Missing values after imputation for Cabin and Embarked:')
print(df_missing_post_all_imputation)
```

```python
df = df.drop_duplicates()

# Print the DataFrame after dropping duplicates
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Exclude non-numeric columns from correlation analysis
numeric_df = df.select_dtypes(include=['number'])

# Summary statistics
df_description = numeric_df.describe()
print('Summary Statistics:')
print(df_description)

# Correlation matrix
corr_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix')
plt.show()

# Distribution of categorical data
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
plt.subplots_adjust(hspace=0.5)

sns.countplot(x='Survived', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Survival Count')

sns.countplot(x='Pclass', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Passenger Class Distribution')

sns.countplot(x='Sex', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Gender Distribution')

sns.countplot(x='Embarked', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Embarkation Point Distribution')
plt.show()


#Conclusions
#The mean age of passengers is approximately 29.7 years, with a standard deviation of 14.5 years, indicating a wide range of ages among passengers.
#The survival rate is about 38.4%, showing that less than half of the passengers survived.
#There is a significant correlation between fare and survival, suggesting that passengers who paid higher fares had a better chance of survival.
#The majority of passengers embarked from Southampton (S), and there were more males than females on board.
#Passenger class also shows a notable impact on survival, with a higher survival rate observed among first-class passengers.
```

```python
# Visualize the distribution of a numerical feature (e.g., age)
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Visualize the count of a categorical feature (e.g., gender)
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', data=df)
plt.title('Count of Passengers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Univariate Analysis
plt.figure(figsize=(10, 6), facecolor='white')
sns.countplot(x='Survived', data=df)
plt.title('Distribution of Survival')
plt.savefig('univariate_survival.png')
plt.show()

# Bivariate Analysis
plt.figure(figsize=(10, 6), facecolor='white')
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.savefig('bivariate_pclass_survival.png')
plt.show()

# Multivariate Analysis
plt.figure(figsize=(10, 6), facecolor='white')
sns.scatterplot(x='Age', y='Fare', hue='Survived', style='Pclass', data=df)
plt.title('Age, Fare, and Survival by Passenger Class')
plt.savefig('multivariate_age_fare_survival.png')
plt.show()
print('Univariate, bivariate, and multivariate visualizations generated.')
```
